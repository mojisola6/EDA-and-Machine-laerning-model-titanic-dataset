# Data Exploration
is the process of analyzing data to bring out hidden patter in the dataset
- uni-variate Analysis: i.e analyzing one column
- Bi-variate Analysis: i.e analyzing two columns
- Multi-variate Analysis: i.e analyzing more than two columns

# Uni-variate Analysis

```python
survived= train["Survived"].value_counts()
survived.index = ["dead", "Survived"]
survived
```

```python
# pandas plot
survived.plot(kind="bar", color=["orange"])
plt.title("count of survived rate", color="red")
plt.xticks(rotation=35, color="red")
plt.yticks(rotation=35, color="red")
plt.show()
```

```python
dead=train.query("Survived==0")["Sex"].value_counts()
alive=train.query("Survived==1")["Sex"].value_counts()
```

```python
pd.concat([dead, alive], keys=["dead", "alive"], axis=1)
```

```python
# join te dead and the alive series together
gender_survived_rate=pd.concat([dead, alive], keys=["dead", "alive"], axis=1)
```

```python
gender_survived_rate.plot(kind="bar")
plt.title("count of survived rate per sex", color="purple")
plt.xticks(rotation=50, color="black")
plt.yticks(rotation=50, color="black")
plt.show()
```

```python
def barplot(col):
    dead=train.query("Survived==0")[col].value_counts()
    alive=train.query("Survived==1")[col].value_counts()
    gender_survived_rate=pd.concat([dead, alive], keys=["dead", "alive"], axis=1)
    gender_survived_rate.plot(kind="bar")
    plt.title("count of survived rate per sex", color="purple")
    plt.xticks(rotation=50, color="black")
    plt.yticks(rotation=50, color="black")
    plt.show()
```
